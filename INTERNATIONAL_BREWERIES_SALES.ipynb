### Introduction
International Breweries Plc manufactures beer and non-alcoholic malt drinks. Its activities include the brewing, packaging, and marketing of beer, alcoholic flavoured and non-flavoured alcoholic beverages, and soft drinks. The company was founded by Lawrence Omole on December 22, 1971 and is headquartered in Lagos, Nigeria.

### Project Aim:
The goal of this project is to analyze sales data from International Breweries using SQL for data extraction and manipulation, and Python for analysis and visualization. The data covers sales across different regions, countries, months, and sales representatives, focusing on various brands. The analysis aims to provide insights into sales performance, profitability, and return on investment (ROI) across different regions, country and brands.


### Dataset Description
The dataset used in this project includes the following columns:

- Sales_ID: Unique identifier for each sales transaction.
- Region: The geographical region where the sale occurred.
- Months: The month of the sale.
- Year: The year of the sale.
- Sales_Rep: The sales representative responsible for the transaction.
- Brands: The brand of the product sold.
- Unit_Cost: The cost of one unit of the product.
- Unit_Price: The selling price of one unit of the product.
- Quantity: The number of units sold.
- Country: The country where the sale took place.
- Total_Cost: The total cost for the quantity sold.
- Total_Sale: The total revenue generated from the sale

## Analyzing the sales performance


### Top 5 sales representatives based on the total quantity sold, and brands they sell the most

```python
query= '''
    WITH TopSalesReps AS (
    SELECT sales_rep, SUM(quantity) AS total_quantity_sold
    FROM int_brew
    GROUP BY sales_rep
    ORDER BY total_quantity_sold DESC
    LIMIT 5
)
    SELECT tsr.sales_rep, ib.brands, SUM(ib.quantity) AS quantity_sold
    FROM TopSalesReps AS tsr
    JOIN int_brew AS ib ON tsr.sales_rep = ib.sales_rep
    GROUP BY tsr.sales_rep, ib.brands
    ORDER BY tsr.sales_rep, brands;
'''
top_sr= pd.read_sql_query(query, conn)
top_sr
```

```python
# Set the figure size
plt.figure(figsize=(14, 8))

# Create a grouped bar chart
sns.barplot(data=top_sr, x='sales_rep', y='quantity_sold', hue='brands')

# Add titles and labels
plt.title('Quantity of Brands Sold by Sales Representatives')
plt.ylabel('Quantity Sold')
plt.xlabel('Sales Representative')
plt.xticks(rotation=45)
plt.legend(title='Brands', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

### Sales representative had the highest sales in each country, and their contribution to the total sales in that country

```python
query= '''
    WITH CountrySalesRep AS(
    SELECT country, sales_rep, SUM(quantity) AS total_sales
    FROM int_brew
    GROUP BY country, sales_rep
)
    SELECT csr.country, csr.sales_rep, csr.total_sales,
       (csr.total_sales / SUM(csr.total_sales) OVER(PARTITION BY csr.country)) * 100 AS contribution_percentage
    FROM CountrySalesRep csr
    ORDER BY csr.country, csr.total_sales DESC;
'''
srcp= pd.read_sql_query(query, conn)
srcp
```

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 18), constrained_layout=True)
fig.suptitle('Top Sales Representatives Contribution by Country', fontsize=16)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Countries for each subplot
countries = srcp['country'].unique()

# Plot each country's data
for i, country in enumerate(countries):
    sns.barplot(x='contribution_percentage', y='sales_rep', data=srcp[srcp['country'] == country], ax=axes[i])
    axes[i].set_title(f'{country}')
    axes[i].set_xlabel('Contribution Percentage')
    axes[i].set_ylabel('Sales Representative')

# Remove the last subplot if there are only 5 subplots needed
if len(countries) < len(axes):
    fig.delaxes(axes[-1])

plt.show()
```

## Brand Analysis

### What is the percentage distribution of sales revenue across the different brands

```python
query='''
    SELECT brands, SUM(total_sale) AS total_sales_revenue,
           (SUM(total_sale) / (SELECT SUM(total_sale) FROM int_brew) * 100) AS revenue_percentage
    FROM int_brew
    GROUP BY brands
    ORDER BY revenue_percentage DESC;
''' 
pdsr= pd.read_sql_query(query, conn)
pdsr
```

### Monthly Trends

```python
query = """
    SELECT 
        year,
        months,
        SUM(total_sale) AS monthly_total_sales
    FROM 
        int_brew
    WHERE 
        year BETWEEN 2018 AND 2020
    GROUP BY 
        year, months
    ORDER BY 
        year, FIELD(months, 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December');
"""
df = pd.read_sql(query, conn)
df
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='months', y='monthly_total_sales', hue='year', marker='o')

plt.title('Monthly Trend of Total Sales (2018-2020)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.grid(True)
plt.tight_layout()

plt.show()
```

### Total sales revenue generated by each sales representative, and brand that  contributed the most to their sales

```python
query= '''
    WITH SalesRepRevenue AS (
        SELECT sales_rep, SUM(total_sale) AS total_sales_revenue
        FROM int_brew
        GROUP BY sales_rep
    )
    SELECT srr.sales_rep, ib.brands, SUM(ib.total_sale) AS brands_contribution
    FROM SalesRepRevenue AS srr
    JOIN int_brew AS ib ON srr.sales_rep = ib.sales_rep
    GROUP BY ib.brands, srr.sales_rep
    ORDER BY srr.total_sales_revenue, brands_contribution DESC;
'''
tsrb = pd.read_sql_query(query, conn)
tsrb
```

```python

plt.figure(figsize=(14, 8))
sns.barplot(data=tsrb, x='sales_rep', y='brands_contribution', hue='brandS')
plt.title('Brands Contribution by Sales Representative')
plt.ylabel('Contribution')
plt.xlabel('Sales Representative')
plt.xticks(rotation=45)
plt.show()
```

#### Average unit price for each brand, and how does it vary across all countries

```python
query = '''
    WITH avg_unit_price AS 
    (SELECT brands, AVG(unit_price) AS avg_unit_price
    FROM  int_brew
    GROUP BY brands
)    
    SELECT aup.brands, ib.country, (aup.avg_unit_price) AS avg_unit_price
    FROM avg_unit_price AS aup
    JOIN int_brew AS ib ON aup.brands = ib.brands
    GROUP BY ib.country,  
    ORDER BY aup.avg_unit_price
'''
Avg_unit_price = pd.read_sql_query(query, conn) 
Avg_unit_price
```
